```python
import sys
sys.path.append("../")
```

```python
from pathlib import Path
```

```python
folder = Path("../no_duplicates/")
```

```python
from preprocess_utils import load_jsonb
brief_words = {}


for file in (f for f in folder.iterdir() if f.suffix == ".jsonb"):
    for br in load_jsonb(file):
        brief = br["brief"]
        if brief == None:
            continue
        before_column = brief.split(":")[0] + ":"
        before_dash = brief.split("-")[0] + "-"
        after_dash = "-" + brief.split("-")[-1]

        for x in [before_column, before_dash, after_dash]:
            if x not in brief_words:
                brief_words[x] = 0
            brief_words[x] += 1

```

```python
for k, v in sorted(brief_words.items(), key=lambda x: x[1], reverse=True):
    if v > 20:
        print(f"{k}: {v}")
```
